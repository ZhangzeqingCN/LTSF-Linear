```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
x = torch.load(f="models.Linear-Input").to('cpu')
x.shape,x.device
```

```python
X = x.permute(0, 2, 1)
```

```python
B, C, P = X.shape
d_k = P // 2
d_q = d_k
d_v = P // 2
```

```python
WQ = nn.Linear(P, d_q)
WK = nn.Linear(P, d_k)
WV = nn.Linear(P, d_v)
Q = WQ(X)
K = WK(X)
V = WV(X)

Q.shape
```

```python
attn = nn.MultiheadAttention(embed_dim=d_q, num_heads=1, kdim=d_k, vdim=d_v)

attn
```

```python
Y, W = attn(Q, K, V)
```
